### What does Web search columns represent in above data frames?
Numbers represent search interest relative to the highest point on the chart for the given region and time. A value of 100 is the peak popularity for the term. A value of 50 means that the term is half as popular. A score of 0 means there was not enough data for this term.

Basically, the actual search volume of a term is not publicly available. Google only offers a scaled number. Each data point is divided by the total searches of the geography and time range it represents to compare relative popularity.

## Visuaisations


```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

### Tesla stock price and search trend 

```python
plt.figure(figsize=(14,8))
plt.title("TSLA Web search vs Price",fontsize=14)
plt.xticks(fontsize=14,rotation=45)

ax1=plt.gca()
ax2=ax1.twinx()

ax1.plot(df_tesla.MONTH,df_tesla.TSLA_USD_CLOSE,color="RED",linewidth=2)
ax2.plot(df_tesla.MONTH,df_tesla.TSLA_WEB_SEARCH,color="GREEN",linewidth=2)

ax1.set_ylabel("TSLA stock price",color="RED",fontsize=14)
ax2.set_ylabel("TSLA web search",color="GREEN",fontsize=14)

ax1.set_xlim([df_tesla.MONTH.min(),df_tesla.MONTH.max()])

## formatting ticks on x-axis
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_minor_locator(mdates.MonthLocator())
ax1.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.title("BTC News search vs Resampled Price",fontsize=14)
plt.xticks(fontsize=14,rotation=45)

ax1=plt.gca()
ax2=ax1.twinx()

ax1.plot(df_btc_search.MONTH,df_btc_search.BTC_NEWS_SEARCH,color="RED",linewidth=2,linestyle="dashed")
ax2.plot(df_btc_price.DATE,df_btc_price.CLOSE,color="GREEN",linewidth=2,marker=".",markersize=5)

ax1.set_ylabel("Bitcoin closing price",color="RED",fontsize=14)
ax2.set_ylabel("Bitcoin related search",color="GREEN",fontsize=14)

ax1.set_xlim([df_btc_search.MONTH.min(),df_btc_search.MONTH.max()])

## formatting ticks on x-axis
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_minor_locator(mdates.MonthLocator())

ax1.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.title("Monthly Search of 'Unemployment Benefits' in the U.S. vs the U/E Rate",fontsize=14)
plt.xticks(fontsize=14,rotation=45)

ax1=plt.gca()
ax2=ax1.twinx()

ax1.plot(df_unemployment.MONTH,df_unemployment.UE_BENEFITS_WEB_SEARCH.rolling(window=3).mean(),color="RED",linewidth=2,linestyle="dashed")
ax2.plot(df_unemployment.MONTH,df_unemployment.UNRATE.rolling(window=6).mean(),color="GREEN",linewidth=2,marker=".",markersize=5)

ax1.set_ylabel("Unemployment Benefits web search",color="RED",fontsize=14)
ax2.set_ylabel("Unemployment rates",color="GREEN",fontsize=14)

ax1.set_xlim([df_unemployment.MONTH.min(),df_unemployment.MONTH.max()])

## formatting ticks on x-axis
ax1.xaxis.set_major_locator(mdates.YearLocator())
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_minor_locator(mdates.MonthLocator())

ax1.grid(axis='y')
plt.show()
```
